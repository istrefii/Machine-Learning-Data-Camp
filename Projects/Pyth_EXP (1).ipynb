```python
# Load the packages

import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

EPL18 = pd.read_csv('international_matches.csv')
print(EPL18.columns.tolist())
```

```python
EPL18.head()
```

```python

EPL18= EPL18[EPL18.tournament != 'Friendly']
```

```python
EPL18=EPL18.rename(columns={'home_team': 'HomeTeam', 'away_team': 'AwayTeam', 'home_team_score': 'FTHG', 'away_team_score': 'FTAG'})
EPL18.head()
```

```python
EPL18['FTR']= np.where(EPL18.home_team_result == 'Win', 'H', np.where(EPL18.home_team_result== 'Draw', 'D', 'A'))
EPL18.head()
```

```python
EPL18= EPL18[['date', 'HomeTeam',	'AwayTeam',	'FTHG',	'FTAG',	'FTR']]
EPL18.head()
```

```python
EPL18['hwin']= np.where(EPL18.FTR=='H', 1, np.where(EPL18.FTR=='D', .5, 0))
EPL18['awin']= np.where(EPL18.FTR=='A', 1, np.where(EPL18.FTR=='D', .5, 0))
EPL18['count']= 1
EPL18.head()
```

```python
df2017= EPL18[EPL18.date > '2018-07-01']
df2017.head()
```

```python
home_2017= df2017.groupby('HomeTeam')['count', 'hwin', 'FTHG', 'FTAG'].sum().reset_index()
home_2017= home_2017.rename(columns={'HomeTeam': 'Team', 'count': 'Ph', 'FTHG': 'FTHGh', 'FTAG': 'FTAGh'})
home_2017.head()
```

```python
away_2017= df2017.groupby('AwayTeam')['count', 'awin', 'FTHG', 'FTAG'].sum().reset_index()
away_2017= away_2017.rename(columns={'AwayTeam': 'Team', 'count': 'Pa', 'FTHG': 'FTHGa', 'FTAG': 'FTAGa'})
away_2017.head()
```

```python
EPL17= pd.merge(home_2017, away_2017, on='Team')
EPL17
```

```python
EPL17['win']=  EPL17.hwin + EPL17.awin
EPL17['G']= EPL17.Ph + EPL17.Pa
EPL17['GF']= EPL17['FTHGh']+EPL17['FTAGa']
EPL17['GA']= EPL17['FTAGh']+EPL17['FTHGa']
EPL17.head(10)
```

```python
EPL17['wpc17']= EPL17.win/EPL17.G
EPL17['pyth17'] = EPL17['GF']**2/(EPL17['GF']**2 + EPL17['GA']**2)
EPL17.head()
```

```python
teams_worldcup= ['Qatar', 'Ecuador', 'Senegal', 'Netherlands', 'England', 'USA',
                  'Wales', 'IR Iran','Argentina', 'Saudi Arabia', 'Mexico', 'Poland', 'France', 
                  'Australia', 'Denmark', 'Tunisia', 'Spain', 'Costa Rica', 'Germany', 
                  'Japan', 'Belgium', 'Canada', 'Morocco', 'Croatia', 'Brazil', 'Serbia', 
                  'Switzerland', 'Cameroon', 'Portugal', 'Ghana', 'Uruguay', 'Korea Republic']
```
